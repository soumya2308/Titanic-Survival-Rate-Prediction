# DATA CLEANING

```python
# lets see is thre any relation between pclass and age, for that lets plot boxplot

plt.figure(figsize=(8,5))

# using this we can increase and decrease the size of the graph

sns.boxplot(x="Pclass", y="Age", data=df)

# through the box plot we can see the avg age in each pclass so that we can replace null values with respective age
# we found that mean or avg age of pclass 1 is around 37
# mean or avg age of pclass 2 is around 29
# mean or avg age of pclass 3 is around 25
# so we will replace the null age of pclass 1 with 37 of pclass 2 with 29 and of pclass 3 with 25
```

```python
# the actul size of the graph is

sns.boxplot(x="Pclass", y="Age", data=df)
```

```python
# we will drop the cabin column as it contains lots of null values 

df.drop("Cabin", axis=1, inplace=True)

# for droping column we write this syntax
```

```python
df
```

```python
sns.heatmap(df.isnull(), yticklabels=False)
# we can see using heat map no null values are present for Cabin column as we droped the column
```

```python
# in this we filled the missing age value with avg values that occured in individual pclass

def input_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 25
    else:
         return Age
```

```python
# here we used the apply function to call the function input_age

df["Age"]=df[["Age", "Pclass"]].apply(input_age,axis=1)
```

```python
df
```

```python
# now in heatmap we have no null value
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
# we have to import linear_model so that we can apply logistic regression 

from sklearn import linear_model
```

```python
# we will also have to import train_test_split in order to train the data with few data sets and to test the data in the remaining one  

from sklearn.model_selection import train_test_split
```

```python
# now we will split the data into 4 parts

x_train, x_test, y_train, y_test= train_test_split(df.drop("Survived", axis=1), df["Survived"], test_size=0.30, random_state=101)
```

```python
# we will create an object of logistic regression
# this object will get trained
reg= linear_model.LogisticRegression()
```

```python
# we fitted in the above ceated object now this will get trained and will be used in prediction

reg.fit(x_train, y_train)
```

```python
# reg object is trained x_train and y_train, now we will pass x_test and see the accuracy with y_test

prediction= reg.predict(x_test)
```

```python
prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test, prediction)
```
